```python
import pandas as pd
url = 'https://media.githubusercontent.com/media/StackExchange/Survey/refs/heads/main/packages/archive/2025/results.csv'
df = pd.read_csv(url, low_memory=False)
df.head(5)
```

```python
url_schema = 'https://media.githubusercontent.com/media/StackExchange/Survey/refs/heads/main/packages/archive/2025/schema.csv'
schema_df = pd.read_csv(url_schema)
schema_df
```

```python
# Т3 №1
total_rows = df.shape[0]
print(f'Кол-во строк в датасете : {total_rows}')

unique_respondents = df['ResponseId'].nunique()
print(f'Кол-во уникальных респондентов : {unique_respondents}')
```

```python
# Т3 №2
my_list = schema_df['qname'].tolist()#Перевел колонку с вопросами в формат листа и теперь буду видеть список вопросов
df.isna().sum()
my_list_set = set(my_list)
df_set = set(df)
result = my_list_set.intersection(df_set)#Пересичение
all_answers = df.dropna(subset=result)#тут переменная где у нас только те кто ответил на все вопросы
print(f'Количество респондетов которые ответили на все вопросы : {all_answers.shape[0]}')


```

```python
# ТЗ № 3
stat_WorkExp = df['WorkExp'].agg([
    'mean',
    'median',
    lambda x: x.dropna().mode()[0] if not x.dropna().empty else None # Использую лямбду, так как .mode() возвращает Series, а .agg() требует скалярное значение. Забираем первый элемент [0].
]).rename(index={'<lambda>':'mode'})# Переименовал таким образом что бы в мини таблице была mode вместо <lambda>
stat_WorkExp
```

```python
# ТЗ № 4 ищем (remote) Remote	10931
df['RemoteWork'].unique()#Проверил какие уникальные значения есть в этой колонке
Remote = df['RemoteWork'].value_counts()['Remote']#посчитал сколько людей работает удаленно
Remote
```

```python
#ТЗ №5 : Определение популярности Python
Code_Langueage = df['LanguageHaveWorkedWith'].str.contains('Python',na=False).sum() # Берем колонку с данными из df, используем текстовый метод .str,выделяем что бы нам выдало только значение Python и что бы не было ошибок указывает что там где NA должно быть False после чего ссумируем ячейки где есть python
python_percent = Code_Langueage / total_rows * 100 #создал переменную где сделал вычесление , сколько людей используют в python  по процентому соотношению
result = round(python_percent,1) # 37.5% респондентов работает в Python
print(f'{result}%')
```

```python
#ТЗ №6 : Анализ путей обучения программированию
LearnCode = df['LearnCode'].str.contains('Online Courses',na=False).sum()#Взял колонку с ответами респондентов на то как они получали своё образование , воспользовался методом .str и выделил что бы мне выдавало только там где это было онлайн,там где не есть пропуски указал Na=False
print(f'Количество респондентов которые обучались програмированию с помощью онлайн курсов : {LearnCode}')
```

```python
#ТЗ №7 : Географический анализ компенсации Python-разработчиков
python_mask = df['LanguageHaveWorkedWith'].str.contains('Python',na=False)#Сделал фильтрацию что бы в дальнейшем вывести в таблице только те строки где есть Python и сохранил в переменную
python_df = df[python_mask]#Создал еще одну переменную и указал что бы в ней были только те строки в нашем дф где респонденты работают с Python
slise_country = df.groupby('Country')['ConvertedCompYearly'].agg(['mean','median'])#Сделал групировку по странам и агрегацию по компенсации для респондентов которые програмируют на Python
slise_country
```

```python
#ТЗ № 8 : Анализ образования наиболее оплачиваемых специалистов
Clear_Convert = df.dropna(subset=['ConvertedCompYearly']).sort_values('ConvertedCompYearly',ascending=False).head(5)['EdLevel']#Так как нам нужно будет вывести ТОП-5 респондентов то забивать Nan средним или медианой нет смысла, просто удаляю их + сортировка от большего к меньшему,вывел первые 5 строк после фильтрации от большего к меньшему и в конце прикрепил колонку с образованием респондентов
Clear_Convert
```

```python
#Тз № 9 : Анализ популярности Python по возрастным категориям
df['Age'].unique()
df_age = df.groupby('Age').size()#Сделал групировку по возрасту и посчитал сколько людей в каждой категории ответили на вопрос.
age_python = python_df.groupby('Age').size()#Сделал групировку по возрасту чтобы понять сколько людей в каждой категории возраста используют Python
age_python_category = round(age_python / df_age * 100,1) #Разделил количество питонистов в каждой возрастной группе на общее число респондентов этой же группы и перевел в проценты
age_python_category.rename('Процент Python')
```

```python
#ТЗ № 10 : Анализ индустрий среди высокооплачиваемых удаленных работников
remote_df = df[df['RemoteWork'] == 'Remote']#Понимаем сколько людей работало удаленно
quan_df = remote_df['ConvertedCompYearly'].quantile(0.75)#Нашел денежную границу (75-й перцентиль), выше которой начинаются топ-25% самых больших зарплат
rich_remote_df =  remote_df[remote_df['ConvertedCompYearly'] >= quan_df]#Фильтруем удаленщиков еще раз и оставляем только тех, у кого зарплата выше этой границы
rich_remote_df['Industry'].value_counts()#Считаем, сколько богатых удаленщиков работает в каждой индустрии , показывает от большего к меньшему
```
